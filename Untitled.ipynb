## Apredendo Pandas

```python
import pandas as pd
```

```python
## Criando um dataframe apartir de um dicionario
vendas = {'data':['15/02/2023', '16/02/2023'],
          'valor': [500, 300],
          'produto':['feijão','Arroz'],
           'qtde': [50,70]}

vendas_df = pd.DataFrame(vendas)
```

```python
display(vendas_df)
```

```python
## Criando um dataframe apartir de uma base de dados

vendas2_df = pd.read_excel('Vendas.xlsx')
```

```python
display(vendas2_df)
```

```python
## Mostrando os 10 primeiros registros
display(vendas2_df.head(10))
```

```python
## Mostrar qtas linhas e qtas colunas a tabela tem
print(vendas2_df.shape)
```

```python
display(vendas2_df.describe())
```

```python
## Pegando uma coluna e transformando em uma serie
produtos = vendas2_df['Produto']
display(produtos)
```

```python
produtos2 = vendas2_df[['Produto','ID Loja']]
display(produtos2)
```

```python
#pegar uma linha especifica
display(vendas2_df.loc[1])
```

```python
#pegando mais de uma linha
display(vendas2_df.loc[1:5])
```

```python
# Pegar linhas que corresponde a uma condicao
vendas_norte_shop = vendas2_df.loc[vendas2_df['ID Loja'] == 'Norte Shopping']
display(vendas_norte_shop)
```

```python
# Pegando varias linhas e colunas
#vendas_norte_shop = vendas2_df.loc[linhas, colunas]
display(vendas2_df.loc[vendas2_df['ID Loja'] == 'Norte Shopping',['ID Loja','Produto','Quantidade']])
```

```python
#Pegando uma celula especifica
print(vendas2_df.loc[1,'Produto'])
```

```python
# Adicionando 1 coluna
# a partir de uma coluna que ja existe
vendas2_df['Comissão'] = vendas2_df['Valor Final'] * 0.05
display(vendas2_df)
```

```python
#Criar uma coluna com valor padrão
#dois pontos significa todas as linhas
vendas2_df.loc[:,'Imposto'] = 0
display(vendas2_df)
```

```python
#Adicionando um complemento de mês no dataframe com os mesmo campos
vendas_dez_df = pd.read_excel('Vendas - Dez.xlsx')

vendas2_df = pd.concat([vendas2_df, vendas_dez_df], axis=0)
display(vendas2_df)
```

```python
#Excluir coluna
vendas2_df = vendas2_df.drop('Imposto', axis=1)
display(vendas2_df)
```

```python
#Excluindo um registro
#axis = 1 = coluna
#axis = 0 = linha

vendas2_df = vendas2_df.drop(0,axis=0)
```

```python
display(vendas2_df)
```

```python
#Deletando linhas e colunas  completamente vazias
vendas2_df = vendas2_df.dropna(how='all')
```

```python
#Deletando todas as colunas que estão vazios
vendas2_df = vendas2_df.dropna(how='all', axis=1)
```

```python
display(vendas2_df)
```

```python
#Deletar linhas que possuem pelo menos 1 valor vazio
vendas2_df = vendas2_df.dropna()
```

```python
# Preencher valores vazios com a media da coluna
vendas2_df['Comissão'] = vendas2_df['Comissão'].fillna(vendas2_df['Comissão'].mean())
display(vendas2_df)
```

```python
#Preencher com o ultimo valor
vendas2_df = vendas2_df.ffill()
display(vendas2_df)
```

```python
#Agrupar produtos vendidos por loja(value counts)
transacoes_loja = vendas2_df['ID Loja'].value_counts()
display(transacoes_loja)
```

```python
#Faturamento de cada produto(group by)
faturamento_produto = vendas2_df.groupby('Produto').sum(numeric_only=True)
display(faturamento_produto)
```

```python
faturamento_produto = vendas2_df[['Produto','Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```

```python
# Mesclando 2 dataframes
# Primeiro importando a base de dados a ser mesclada

gerentes_df = pd.read_excel('Gerentes.xlsx')
display(gerentes_df)
```

```python
vendas_df = vendas2_df.merge(gerentes_df)
display(vendas2_df)
```
